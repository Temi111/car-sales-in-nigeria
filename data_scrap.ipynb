```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
Price = []   #Prices for the 1st page
page_link = 'https://www.cheki.com.ng/vehicles'

page_response = requests.get(page_link)
page_content = BeautifulSoup(page_response.content, "html.parser")

#h2 tags holds the price
for i in range(0, 20):
    prices = page_content.find_all("h2")[i].text
    if prices != 'Cheki Finance x':
        Price.append(prices)
```

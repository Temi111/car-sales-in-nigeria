```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
#initialize a dataframe
cars = pd.DataFrame()
```

```python
details = []   #Prices for the 1st page
page_link =  'https://www.cheki.com.ng/vehicle/mercedes-benz-glk-350-9neejj'

page_response = requests.get(page_link)
page_content = BeautifulSoup(page_response.content, "html.parser")

for i in range(0,10):
    detail = page_content.find_all("dd")[i].text
    details.append(detail)
    
#car-price
for k in range(9,11):
    price = page_content.find_all("span")[k].text
    details.append(price)
    

#car name
Name = page_content.h1.text
details.append(Name)

for j in range(59,82):
    look = page_content.find_all("li")[j].text
    details.append(look)

#Remove the too-long description here
details = details[1:]

#Append to the dataframe
cars = cars.append(pd.Series(details), ignore_index=True)

#save to disk
cars.to_csv('cars_final_2.csv')
```
